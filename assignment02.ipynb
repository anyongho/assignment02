## import packages for plotting graphs and manipulate data

```python
import numpy as np
import matplotlib.pyplot as plt
```

## define my fuction : 
## f(x) = x + sin(x)

```python
def myFunction(x):
    func = np.sin(x) + x
    return func
```

## define the derivative my function : 
## f'(X) = 1 + cos(x)

```python
def myDerivativeFunction(x):
    Deriv = 1 + np.cos(x)
    return Deriv
```

## pick 3 points of the domain (-1.57, 0,  3.14)  Ï€ = 3.14

```python
def newfunc(x):
    new = myDerivativeFunction(0) * x 
    return new

def newfunc2(x):
    new = myDerivativeFunction(3.14) * x + 2.57
    return new

def newfunc3(x):
    new = myDerivativeFunction(-1.57) * x - 1
    return new
```

## define the domain:
## x = [-10 : 0.1 : 10]

```python
x = np.arange(-10, 10, 0.1)
```

## compute the graph

```python
f = myFunction(x)
Df = myDerivativeFunction(x)
```

## plot the graphs for function and derivative function

```python
plt.figure(1)
plt.plot(x, f, 'b', label="function")
plt.plot(x, Df, 'r', label="derivative")
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.)
plt.show()
```

## plot the graphs for 3 points

```python
plt.figure(2)
plt.plot(x, newfunc(x), 'r', label="0 point")
plt.plot(x, newfunc2(x), 'r', label="3.14 point")
plt.plot(x, newfunc3(x), 'r', label="-1.57 point")
plt.legend(bbox_to_anchor=(1.1, 1), loc=2, borderaxespad=0.)
plt.show()
```
